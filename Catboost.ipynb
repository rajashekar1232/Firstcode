```python
import ccxt
import pandas as pd
from datetime import datetime, timedelta
import time

# Initialize the Binance exchange
exchange = ccxt.binance({
    'rateLimit': 1200,
    'enableRateLimit': True,
})

# Function to fetch OHLCV data in chunks
def fetch_ohlcv(symbol, timeframe, since, limit=1000):
    all_data = []
    while True:
        try:
            data = exchange.fetch_ohlcv(symbol, timeframe, since=since, limit=limit)
        except Exception as e:
            print(f"Error fetching data: {e}")
            time.sleep(5)  # Wait before retrying
            continue
        if not data:
            break
        all_data.extend(data)
        since = data[-1][0] + 1  # Increment timestamp to fetch the next batch
        print(f"Fetched {len(data)} rows. Total: {len(all_data)}")
        if len(data) < limit:  # If fewer rows returned, we've reached the end
            break
        time.sleep(1)  # Respect Binance's rate limits
    return all_data

# Parameters
symbol = 'BTC/USDT'  # Binance uses USDT as a stablecoin
timeframe = '15m'  # 15-minute interval
end_date = datetime.utcnow()
start_date = end_date - timedelta(days=2*365)  # 2 years of data
since = int(start_date.timestamp() * 1000)  # Convert to milliseconds

# Fetch OHLCV data
print("Fetching OHLCV data... This may take a while.")
ohlcv_data = fetch_ohlcv(symbol, timeframe, since)

# Convert data to a DataFrame
columns = ['timestamp', 'open', 'high', 'low', 'close', 'volume']
df = pd.DataFrame(ohlcv_data, columns=columns)
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')

# Save to CSV
filename = f"{symbol.replace('/', '_')}_15m_ohlcv.csv"
df.to_csv(filename, index=False)
print(f"Data saved to {filename}")
```

```python
import ccxt
import pandas as pd
from datetime import datetime, timedelta
import time

# Initialize the Binance exchange using CCXT
exchange = ccxt.binance()

# Define symbol, timeframe, and date range
symbol = 'BTC/USDT'  # Binance uses BTC/USDT for spot market
timeframe = '15m'

# Define the date range (2 years ago to now)
end_datetime = datetime.utcnow()
start_datetime = end_datetime - timedelta(days=730)

# Define the limit for the number of data points per request
limit = 1000  # Binance allows up to 1000 data points per request

# Initialize the list to store data
all_data = []

while start_datetime < end_datetime:
    since = int(start_datetime.timestamp() * 1000)

    try:
        ohlcv = exchange.fetch_ohlcv(symbol, timeframe, since=since, limit=limit)
    except ccxt.BaseError as e:
        print(f"An error occurred: {e}")
        time.sleep(5)
        continue

    if not ohlcv:
        break

    all_data.extend(ohlcv)
    last_timestamp = ohlcv[-1][0]
    start_datetime = datetime.utcfromtimestamp(last_timestamp / 1000) + timedelta(minutes=15)
    print(f"Fetched {len(ohlcv)} rows. Next start time: {start_datetime}")

    time.sleep(exchange.rateLimit / 1000)

df = pd.DataFrame(all_data, columns=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume'])
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='ms')
df.set_index('Timestamp', inplace=True)
df.to_csv("BTC_USDT_15m_binance.csv")
print("Data saved to BTC_USDT_15m_binance.csv")
```
